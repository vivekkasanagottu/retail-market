```python
conversion_rates = df.groupby('Dealer Name').agg({'Enquiry Id': 'count',' Order Recd\Lost ': lambda x: sum(x == 'Order Received')})
conversion_rates['Conversion Rate'] = conversion_rates[' Order Recd\Lost '] / conversion_rates['Enquiry Id']
```

```python
conversion_rates
```

```python
threshold = 0.15
performance_levels = np.where(conversion_rates['Conversion Rate'] >= threshold, 'Good Performer', 'Bad Performer')
conversion_rates['Performance Level'] = performance_levels
```

```python
conversion_rates
```

```python
import matplotlib.pyplot as plt

plt.scatter(conversion_rates['Enquiry Id'], conversion_rates['Conversion Rate'], c=performance_levels == 'Bad Performer')
plt.xlabel('Number of Enquiries')
plt.ylabel('Conversion Rate')
plt.show()
```

```python
plt.scatter(conversion_rates['Enquiry Id'], conversion_rates['Conversion Rate'], c=performance_levels == 'Good Performer')
plt.xlabel('Number of Enquiries')
plt.ylabel('Conversion Rate')
plt.show()
```
